```python
import pandas as pd

df = pd.read_csv('student.csv')

# Mendapatkan statistik deskriptif untuk atribut numerik
numeric_columns = df.select_dtypes(include=['number'])
descriptive_stats = numeric_columns.describe()

# Menambahkan kuartil (Q1, Q2, Q3) ke statistik deskriptif
quartiles = numeric_columns.quantile([0.25, 0.5, 0.75])
# Use concat instead of append and ignore index to avoid duplicate indexes
descriptive_stats = pd.concat([descriptive_stats, quartiles], ignore_index=True)

# Mengganti nama indeks kuartil
descriptive_stats = descriptive_stats.rename(index={0.25: 'Q1', 0.5: 'Q2', 0.75: 'Q3'})

# Menampilkan hasil
print(descriptive_stats)
```

```python
import pandas as pd
df = pd.read_csv('student.csv')

# Jumlah record
num_records = df.shape[0]

# Jumlah attribute
num_attributes = df.shape[1]

# Nama attribute
attribute_names = df.columns.tolist()

# Jumlah attribute dengan tipe angka
num_numeric_attributes = df.select_dtypes(include=['number']).shape[1]

# Menampilkan informasi
print("Jumlah record:", num_records)
print("Jumlah attribute:", num_attributes)
print("Nama attribute:", attribute_names)
print("Jumlah attribute dengan tipe angka:", num_numeric_attributes)
```

```python
import pandas as pd
df = pd.read_csv('student.csv')
# Memilih kolom object
object_columns = df.select_dtypes(include=['object'])

# Menghitung jumlah label unik untuk setiap kolom object
for column in object_columns:
    num_labels = df[column].nunique()
    print(f"Kolom '{column}' memiliki {num_labels} label unik.")
```
